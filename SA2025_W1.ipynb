### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
import pandas as pd

# Load the dataset
file_path = 'Cars.csv'
df = pd.read_csv(file_path)

# Convert horsepower to numeric, coercing errors to NaN
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Filter cars with horsepower > 100 and weight < 3000
filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Find the most common origin
most_common_origin = filtered_df['origin'].mode()

print(most_common_origin)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
import pandas as pd

# Load dataset
df = pd.read_csv('Cars.csv')

# Convert horsepower to numeric, coercing errors to NaN
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Drop rows with missing horsepower or weight
df = df.dropna(subset=['horsepower', 'weight'])

# Calculate median mpg
median_mpg = df['mpg'].median()

# Filter cars with mpg above median
filtered_df = df[df['mpg'] > median_mpg].copy()

# Create horsepower to weight ratio column
filtered_df['hp_per_weight'] = filtered_df['horsepower'] / filtered_df['weight']

# Find max ratio
max_ratio = filtered_df['hp_per_weight'].max()

# Find cars with max ratio
best_cars = filtered_df[filtered_df['hp_per_weight'] == max_ratio]

# Display relevant columns
print(best_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_per_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Cars.csv')

# Group by model_year and origin, then calculate mean mpg
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Set the plot style
sns.set(style='whitegrid')

# Create the multi-line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Cars.csv')

# Ensure correct data types and handle missing values
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df = df.dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])

# Set the desired hue order
hue_order = ['japan', 'europe', 'usa']

plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(40, 400),
    alpha=0.7,
    hue_order=hue_order,
    palette='Set2'
)

plt.title('Horsepower vs Weight of Cars\nColor by Origin, Size by MPG', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight (lbs)', fontsize=14)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd

# Load the dataset
df = pd.read_csv('Cars.csv')

# Ensure correct data types
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Find car names with more than one unique model_year
car_counts = df.groupby('name')['model_year'].nunique()
cars_multiple_years = car_counts[car_counts > 1].index

# Filter for these cars
filtered_df = df[df['name'].isin(cars_multiple_years)]

# Compute std(mpg) for each car name
mpg_std = filtered_df.groupby('name')['mpg'].std()

# Filter for std(mpg) < 1.0
consistent_cars = mpg_std[(mpg_std < 1.0)]

# Get appearance counts for these cars
consistent_car_counts = df[df['name'].isin(consistent_cars.index)].groupby('name').size()

# Keep only those with at least 2 appearances
consistent_car_counts = consistent_car_counts[consistent_car_counts >= 2]
consistent_cars = consistent_cars[consistent_cars.index.isin(consistent_car_counts.index)]


# Aggregate results for consistent cars
result_df = df[df['name'].isin(consistent_cars.index)].groupby('name').agg(
    appearances=('name', 'size'),
    avg_mpg=('mpg', 'mean')
)
result_df['std_mpg'] = consistent_cars

# Sort: number of appearances (descending), then avg mpg (descending)
result_df_sorted = result_df.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

print(result_df_sorted)

```
